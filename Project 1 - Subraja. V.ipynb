```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore') #ignoring repeated and unnecessary warnings.
```

```python
columns_name=['user_id','item_id','rating','timestamp'] #naming the columns of the table obtained from csv file.
df=pd.read_csv('u.data',sep="\t",names=columns_name) #reading the table from csv file, 'u.data'.
df.head() #displaying the first/top 5 entries. 
```

```python
df['user_id'].nunique() #the number of unique user IDs.
```

```python
df['item_id'].nunique() #the number of distinct movies.
```

```python
movies_title=pd.read_csv('u.item',sep="\|",header=None,encoding="ISO-8859-1") #reading the table from csv file, 'u.item'.
movies_title.head()
```

```python
movies_titles=movies_title[[0,1]] #removing the unnecessary columns, and keeping only the first 2 columns.
movies_titles.columns=['item_id','title'] #naming the columns.
movies_titles.head()
```

```python
df=pd.merge(df,movies_titles,on='item_id') #merging the tables by grouping their item IDs.
df
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating']) #creating a 'ratings' table, by grouping the title, and appending the mean of ratings of each movie. 
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title').count()['rating']) #adding another column, 'num of ratings' using count() function, to the 'ratings' table.
ratings.head()
```

RECOMMENDATION SYSTEM:

```python
moviemat=df.pivot_table(index='user_id',columns='title',values='rating') #creating a pivot table with user ID, movies, and rating of each user on each movie.
moviemat.head()
```

```python
def predict_movies(movie_name): #a function which returns our required output i.e. movie recommendations.
    movie_user_ratings=moviemat[movie_name]
    similar_to_movie=moviemat.corrwith(movie_user_ratings) #corrwith() function correlates 2 attributes.
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlation']) #creating a 'correlation' table.
    corr_movie.dropna(inplace=True) #dropping all the NaN values.
    corr_movie=corr_movie.join(ratings['num of ratings']) #merges the correlation table with 'num of ratings'
    
    predictions=corr_movie[corr_movie['num of ratings']>100].sort_values('correlation',ascending=False) #considers only the movies which got atleast 100 ratings, correlates it with the movie that's inputted and sorts it in Descending order.
    
    return predictions
```

```python
predict_my_movie=predict_movies('Star Wars (1977)') #calls the 'predict_movies' function.
predict_my_movie.head() #returns top 5 movie recommendations, based on what you search for.
```

```python
predict_my_movie=predict_movies('Home Alone (1990)')
predict_my_movie.head(10) #returns top 10 movie recommendations, based on what you search for.
```
